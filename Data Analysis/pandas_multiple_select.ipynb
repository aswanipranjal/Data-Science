```python
import pandas as pd
ufo = pd.read_csv('http://bit.ly/uforeports')
ufo.head(3)
```

```python
# loc is filtering rows and columns by label
# returns row 0, all columns
ufo.loc[0, :]
```

```python
ufo.loc[[0, 1, 2], :]
```

```python
ufo.loc[:2, :]
```

```python
# loc parameters are inclusive on both sides. It doesn't exclude the index after the colon
ufo.loc[:, 'City']
```

```python
ufo.loc[:, 'City':'State']
```

```python
ufo.loc[:2, ['City', 'State']]
```

```python
ufo.loc[:2, 'City':'State']
```

```python
# Same as
ufo.head(3).drop('Time', axis=1)
```
