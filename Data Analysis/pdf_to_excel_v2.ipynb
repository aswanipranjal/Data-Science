```python
import pandas as pd
df = pd.DataFrame()
data = dict()
data['Name'] = []
data['Email'] = []
data['Mobile'] = []
data['Branch'] = []
```

```python
def scrape(filepath):
    pdf = pdfquery.PDFQuery(filepath)
    pdf.load(0)
    label = pdf.pq('LTTextLineHorizontal:contains("Personal Details")')
    left_corner = float(label.attr('x0'))
    bottom_corner = float(label.attr('y0'))
    name = pdf.pq('LTTextLineHorizontal:in_bbox("%s, %s, %s, %s")'%(left_corner, bottom_corner-20, left_corner+250, bottom_corner)).text()
#     name
    data['Name'].append(name)
    pdf.load(1)
    label = pdf.pq('LTTextLineHorizontal:contains("Editable Details")')
    label.text()
    lc = float(label.attr('x0'))
    bc = float(label.attr('y0'))
    email = pdf.pq('LTTextLineHorizontal:in_bbox("%s, %s, %s, %s")'%(lc+20, bc-20, lc+300, bc)).text()
    
    data['Email'].append(email)
    mobile = pdf.pq('LTTextLineHorizontal:in_bbox("%s, %s, %s, %s")'%(lc+20, bc-40, lc+300, bc-20)).text()
    
    data['Mobile'].append(mobile)
    label = pdf.pq('LTTextLineHorizontal:contains("Bank Details")')
#     label.text()
    lc = float(label.attr('x0'))
    bc = float(label.attr('y0'))
    branch = pdf.pq('LTTextLineHorizontal:in_bbox("%s, %s, %s, %s")'%(lc+20, bc+0, lc+280, bc+60)).text()
    
    data['Branch'].append(branch)
#     print(data)
#     row = pd.Series(data)
#     print(row)
#     df.append(row, ignore_index=True)

# for i in range(400, 1000):
#     filepath = '2016 batch/Student_profile_2012JE0' + str(i) + '.pdf'
# #     print(filepath)
#     try:
#         scrape(filepath)
#     except FileNotFoundError:
#         print("Skipped " + str(i))
#         pass
    
for i in range(1000, 1411):
    filepath = '2016 batch/Student_profile_2012JE' + str(i) + '.pdf'
    try:
        scrape(filepath)
    except FileNotFoundError:
        print("Skipped " + str(i))
        pass
```
