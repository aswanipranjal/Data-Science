```python
# To only read an excel file without separating into sheets
df = pd.read_excel('http://pythonprogramming.net/static/downloads/machine-learning-data/titanic.xls')
df.head()
```

```python
# To convert non-numeric data into numeric data, 
# we take the set of the sex column and map the keys to integers
df.drop(['body', 'name'], axis=1, inplace=True)
```

```python
df.head()
df.dtypes
```

```python
df.convert_objects(convert_numeric=True)
df.head()
```

```python
df.dtypes
```

```python
# Fill missing data with zeros
df.fillna(0, inplace=True)
df.head()
```

```python
def handle_non_numerical_data(df):
    columns = df.columns.values
    
    for column in columns:
        text_digit_vals = {}
        def convert_to_int(val):
            return text_digit_vals[val]
        
        if df[column].dtype != np.int64 and df[column].dtype != np.float64:
            column_contents = df[column].values.tolist()
            unique_elements = set(column_contents)
            x = 0
            for unique in unique_elements:
                if unique not in text_digit_vals:
                    text_digit_vals[unique] = x
                    x += 1
                    
            df[column] = list(map(convert_to_int, df[column]))
            
    return df

df = handle_non_numerical_data(df)
df.head()
```

## Clustering
### <i> K-Means </i> clustering on the titanic dataset

```python
X = np.array(df.drop(['survived'], axis=1).astype(float))
y = np.array(df['survived'])
```

```python
X, y
```

```python
clf = KMeans(n_clusters=2)
clf.fit(X)
```

```python
y_pred = clf.predict(X)
from sklearn import metrics
print(metrics.accuracy_score(y, y_pred))
```

```python
correct = 0
for i in range(len(X)):
    to_predict = np.array(X[i].astype(float))
    to_predict = to_predict.reshape(-1, len(to_predict))
    predictions = clf.predict(to_predict)
    if predictions[0] == y[i]:
        correct += 1
        
print(correct/len(X))
```

```python
# So we get about 51% accuracy on the titanic training dataset
# without any preprocessing
# We can scale (normalize) the data before fitting the classifier
from sklearn import preprocessing
X = preprocessing.scale(X)
clf.fit(X)
y_pred = clf.predict(X)
print(metrics.accuracy_score(y, y_pred))
```

```python
X = preprocessing.scale(X)
clf.fit(X)
y_pred = clf.predict(X)
print(metrics.accuracy_score(y, y_pred))
```
