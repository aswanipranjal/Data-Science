```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import style
style.use('ggplot')
```

```python
df = pd.read_csv('Pokemon.csv')
```

```python
df.head(20)
```

```python
# Change columns to upper case
df.columns = df.columns.str.upper().str.replace('_', '')
```

```python
df[df['LEGENDARY']==True].head(10)
```

```python
df = df.set_index('NAME')
```

```python
# Cleaning the dataframe
# Replacing the text before Mega with an empty string
df.index = df.index.str.replace(".*(?=Mega)", "")
```

## Visualizations

```python
bins = range(0, 200, 5)
%matplotlib inline
plt.hist(df['ATTACK'], bins, histtype='bar', rwidth=1.2, color='#dddddd')
```

```python
plt.hist(df['ATTACK'], bins, histtype='bar', rwidth=1.2, color='#333333')
plt.xlabel('Attack')
plt.ylabel('Count')
plt.plot()
plt.axvline(df['ATTACK'].mean(), linestyle='--', color='k')
plt.show()
```
