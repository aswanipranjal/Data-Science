## Visualizations

```python
bins = range(0, 200, 5)
%matplotlib inline
plt.hist(df['ATTACK'], bins, histtype='bar', rwidth=1.2, color='#dddddd')
```

```python
plt.hist(df['ATTACK'], bins, histtype='bar', rwidth=1.2, color='#333333')
plt.xlabel('Attack')
plt.ylabel('Count')
plt.plot()
plt.axvline(df['ATTACK'].mean(), linestyle='--', color='k')
plt.show()
```

```python
df.ATTACK.mean()
```

```python
fire = df[(df['TYPE 1']=='Fire') | (df['TYPE 2']=='Fire')]
```

```python
water = df[(df['TYPE 1']=='Water') | (df['TYPE 2']=='Water')]
```

```python
plt.scatter(fire.ATTACK, fire.DEFENSE, color='r', label='Fire', marker='.', s=50)
plt.scatter(water.ATTACK, water.DEFENSE, color='b', label='Water', marker='.', s=50)
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.legend()
plt.plot()
fig = plt.gcf()
fig.set_size_inches(12, 6)
plt.show()
```
