```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pkmn = pd.read_csv('Pokemon.csv')
```

```python
pkmn.head()
```

```python
pkmn = pkmn.drop(['Generation', 'Legendary'], axis=1)
```

```python
pkmn.head()
```

```python
%matplotlib inline
sns.jointplot(x='HP', y='Attack', data=pkmn)
```

```python
sns.boxplot(y='HP', data=pkmn)
```

```python
sns.boxplot(data=pkmn)
```

```python
sns.violinplot(data=pkmn)
```

```python
pkmn.drop(['Total', '#'], 1, inplace=True)
```

```python
sns.violinplot(data=pkmn)
```

```python
# Trying to use the swarmplot function with the 'hue' argument is going to give us some errors if we don't transform our data a bit though.
# We will use the Pandas' melt function here
pkmn = pd.melt(pkmn, id_vars=['Name', 'Type 1', 'Type 2'], var_name='Stat')
```

```python
pkmn.head(20)
```

```python
pkmn.shape
```

```python
pkmn.tail(20)
```

```python
pkmn.iloc[1000:1100, :]
```

```python
sns.swarmplot(x='Stat', y='value', data=pkmn, hue='Type 1');
```

```python
plt.figure(figsize=(12, 10))
plt.ylim(0, 275)
sns.swarmplot(x='Stat', y='value', data=pkmn, hue='Type 1', split=True, size=7)
plt.legend(bbox_to_anchor=(1, 1), loc=2, borderaxespad=0);
```

```python
sns.set_style('whitegrid') 
with sns.color_palette(['#8ED752', '#F95643', '#53AFFE', '#C3D221', '#BBBDAF', '#AD5CA2', '#F8E64E', '#F0CA42', '#F9AEFE', '#A35449', '#FB61B4', '#CDBD72', '#7673DA', '#66EBFF', '#8B76FF', '#8E6856', '#C3C1D7', '#75A4F9'], n_colors=18, desat=.9):
    plt.figure(figsize=(20, 15))
    plt.ylim(0, 275)
    sns.swarmplot(x='Stat', y='value', data=pkmn, hue='Type 1', split=True, size=7)
    plt.legend(bbox_to_anchor=(1, 1), loc=2, borderaxespad=0);
```
